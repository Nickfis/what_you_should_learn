## getting all data

```python
youtube = build('youtube', 'v3', developerKey=API_KEY)
```

```python
# Split the video_ids list into chunks of 50
video_ids = list(all_videos['video_ids'].unique())
video_id_chunks = [video_ids[i:i + 50] for i in range(0, len(video_ids), 50)]
```

```python
# Split the video_aids list into chunks of 50
video_ids = list(all_videos['video_ids'].unique())
video_id_chunks = [video_ids[i:i + 50] for i in range(0, len(video_ids), 50)]

# Initialize an empty list to hold the data
data = []
counter = 1

for chunk in video_id_chunks:
    print(f"Currently on iteration {counter}")
    # Join the video IDs with commas
    ids = ','.join(chunk)
    request = youtube.videos().list(
        part="snippet,statistics",
        id=ids
    )

    response = request.execute()

    for item in response['items']:
        video_id = item['id']
        pub_date = item['snippet']['publishedAt']
        view_count = item['statistics']['viewCount']
        title = item['snippet']['title']

        # Append the data to the list
        data.append({'Video ID': video_id, 'Publishing Date': pub_date, 
                     'View Count': view_count, 'Title': title})
    counter += 1
```

```python
df = pd.DataFrame(data)
df
```

## Add missing data
All videos that could not be found by their thumbnail have now been getting the correct video id added.
We can now make another request with the video id to the youtube API.

```python
missing_videos = all_videos.loc[~all_videos['video_ids'].isin(df['Video ID'])]['video_ids'].unique()
```

```python
ids = ','.join(missing_videos)
request = youtube.videos().list(
    part="snippet,statistics",
    id=ids
)

response = request.execute()

for item in response['items']:
    video_id = item['id']
    pub_date = item['snippet']['publishedAt']
    view_count = item['statistics']['viewCount']
    title = item['snippet']['title']

    # Append the data to the list
    data.append({'Video ID': video_id, 'Publishing Date': pub_date, 
                 'View Count': view_count, 'Title': title})
```

```python
df = pd.DataFrame(data)
df.rename(columns={'Title': "title_on_youtube"}, inplace=True)
df.columns = ["_".join(col.lower().split(" ")) for col in df.columns]
df.to_csv("video_viewcount.csv", index=False)
```
