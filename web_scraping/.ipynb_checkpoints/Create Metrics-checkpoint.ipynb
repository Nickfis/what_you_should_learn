## How could a metric look like for each course?

## Difficulty
-> Easiest approach: how many people are still watching the last lecture?
Probably not a good measure though. I probably want to have the average loss from video to video and then compare it across.\

-> Middle approach: By how much do we see a decrease in each video? Maybe there are even courses where a subsequent video gains viewers.\

-> Hard approach: 
Try to predict based on the viewing numbers on the video before as well as the first video of the series, how many viewers we'd expect. If above or below, we'll know people find it easier/harder to stick with a given course.

## Predict Watchers for each Video
Model Input:

- First Video Views
- Number of videos for the course
- Video number itself
- Days passed since video was published
- Retention change in the previous video.
- Previous video view count
\

Model Output:
- Number of people watching (true value: `view_count` column)

```python
courses.head()
```

```python
df = (videos[['video_number', 'course_id', 'first_video_views', 'view_count', 'days_since_published']]
      .merge(courses[['id', 'available_videos']], left_on=['course_id'], right_on=['id'])
     .drop(['id'], axis=1))

df['previous_video_view_count'] = df.sort_values(['course_id', 
                                                  'video_number']).groupby('course_id')['view_count'].shift(1)
# how many viewers were kept from the previous video?
# This metric will be dropped again because it would be predictive.
df['retention_from_previous_video'] = df['view_count'] / df['previous_video_view_count']
# However, we can say how many viewers the previous video retained.
df['previous_viewer_retention'] = df.groupby('course_id')['retention_from_previous_video'].shift(1).fillna(-1)
df.drop(['retention_from_previous_video'], axis=1, inplace=True)
df
```

```python
prep_df = df.loc[df['video_number'] > 1]
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from tqdm import tqdm

# Initialize the LeaveOneOut object
loo = LeaveOneOut()

# Apply log transformation to the target
prep_df['log_view_count'] = np.log1p(prep_df['view_count'])

# Define features and target
features = prep_df[['video_number', 'first_video_views', 'available_videos', 'days_since_published',
                   'previous_video_view_count', 'previous_viewer_retention']]
target = prep_df['log_view_count']

# Create a new DataFrame to store results
results = pd.DataFrame(columns=['True Values', 'Predictions'])

# Iterate over each train-test split
for train_index, test_index in tqdm(loo.split(features), total=features.shape[0]):
    # Split the data
    X_train, X_test = features.iloc[train_index], features.iloc[test_index]
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]

    # Create and train the model
    model = RandomForestRegressor(n_estimators=100)
    model.fit(X_train, y_train)

    # Make predictions
    predictions = model.predict(X_test)

    # Transform the predictions back to original scale
    predictions = np.expm1(predictions)
    true_values = np.expm1(y_test.values)

    # Append the true values and predictions to the results DataFrame
    new_row = pd.DataFrame({'True Values': [true_values[0]], 'Predictions': [predictions[0]]})
    results = pd.concat([results, new_row], ignore_index=True)

print(results)
```

```python
final_df = prep_df.reset_index().join(results).drop(['index', 'True Values'], axis=1)
```

```python
final_df = prep_df.reset_index().join(results).drop(['index', 'True Values'], axis=1)
final_df['abs_error'] = abs(final_df['Predictions'] - final_df['view_count'])
final_df['error'] = final_df['view_count'] - final_df['Predictions']

plt.scatter(final_df['Predictions']/1000, final_df['abs_error']/1000)
plt.xlabel("Predictions")
plt.ylabel("Errors");
```

## popularity metric
Expected vs. Actual Views & Average of all videos compared to the last two videos of the course

```python
import pandas as pd
import numpy as np

# Custom function to calculate average excluding the last 2 elements
def avg_without_last_two(group):
    return group[:-2].mean()

# Custom function to calculate average of the last 2 elements
def avg_of_last_two(group):
    return group[-2:].mean()

# Group by 'course_id' and apply the custom functions
average_without_last_two = df.groupby('course_id', as_index=False)['view_count'].apply(avg_without_last_two)
average_of_last_two = df.groupby('course_id', as_index=False)['view_count'].apply(avg_of_last_two)
```

```python
pop_metric = (average_of_last_two.rename(columns={'view_count': "view_count_last_two"})
.merge(average_without_last_two.rename(columns={'view_count': 'view_count_without_last_two'})
                                       , on='course_id')
.assign(last_two_vs_rest = lambda df: df['view_count_last_two'] / df['view_count_without_last_two']))
```

```python
pop_metric.sort_values("last_two_vs_rest", ascending=False)
```

```python
pop_metric = pop_metric.merge(course_popularity[['course_id', 'expected_vs_actual_views']])
pop_metric = pop_metric.merge(courses[['id', 'available_videos']], left_on='course_id', right_on='id')
```

```python
pop_metric
```

```python
pop_metric['standardized_last_two_vs_rest'] = ((pop_metric['last_two_vs_rest'] - pop_metric['last_two_vs_rest'].mean()) 
                                               / pop_metric['last_two_vs_rest'].std())

pop_metric['standardized_expected_vs_actual_views'] = ((pop_metric['expected_vs_actual_views'] - pop_metric['expected_vs_actual_views'].mean()) 
                                               / pop_metric['expected_vs_actual_views'].std())
```

```python
pop_metric['metric'] = pop_metric[['standardized_last_two_vs_rest',
                                   'standardized_expected_vs_actual_views']].mean(axis=1)
pop_metric['metric_pct'] = pop_metric['metric'].rank(pct=True)

pop_metric.rename(columns={'metric': 'popularity_metric',
                          'metric_pct': 'popularity_metric_pct'}, inplace=True)
```
